```python
import cdpam

# between StyleGen v2 and Riffusion

loss_fn = cdpam.CDPAM()
wav_ref = cdpam.load_audio(f'/content/drive/MyDrive/stylegen_paper_riffusion/generated_audio/riffusion_sample.wav')
wav_out = cdpam.load_audio(f'/content/drive/MyDrive/stylegen_paper_v2_results/generated_audio/stylegen_v2_sample.wav')

dist = loss_fn.forward(wav_ref,wav_out)
```
